```python
import pandas as pd
df = pd.read_csv("D:\\My Real Hero\\Haidar03.csv",names=['sl','sw','pl','pw','class'])
# Covert the last column to numaric
df['class']=df['class'].map({"Iris-setosa":0,"Iris-versicolor":1,"Iris-virginica":2})
# Seperate the dataframe in to input and out
x = df.drop(columns=['class'])  # input
y = df['class'] 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
# create a model KNM   (k nearest neighbors model)
from sklearn.neighbors import KNeighborsClassifier
K=KNeighborsClassifier(n_neighbors=5)
# When the model by using training data (80%)
K.fit(x_train,y_train)
# test the model
y_pred_knn=K.predict(x_test)
# find accuracy of model
from sklearn.metrics import accuracy_score
acc_knn=accuracy_score(y_test,y_pred_knn)
acc_knn=round(acc_knn*100,2)
print("accuracy of KNN model is",acc_knn,"%")
```

```python
# imlements logistic regression
import warnings
warnings.filterwarnings('ignore')
from sklearn.linear_model import LogisticRegression
L=LogisticRegression()
L.fit(x_train,y_train)
y_pred_lg= L.predict(x_test)
from sklearn.metrics import accuracy_score
acc_lg = accuracy_score(y_test,y_pred_lg)
acc_lg=round(acc_lg*100,2)
print("Accuracy of logistic reg model is",acc_lg,"%")
```

```python
# implement decision tree model
from sklearn.tree import DecisionTreeClassifier
D= DecisionTreeClassifier()
D.fit(x_train,y_train)
y_pred_dt=D.predict(x_test)
from sklearn.metrics import accuracy_score
acc_dt=accuracy_score(y_test,y_pred_dt)
acc_dt=round(acc_dt*100,2)
print("accuracy of decision tree model is",acc_dt,"%")
```

```python
# implement Naive nayes model
from sklearn.naive_bayes import GaussianNB
N=GaussianNB()
N.fit(x_train,y_train)
y_pred_nb=N.predict(x_test)
from sklearn.metrics import accuracy_score
acc_nb=accuracy_score(y_test,y_pred_nb)
acc_nb=round(acc_nb*100,2)
print("accuracy of naive bayes model is",acc_nb,"%")
```

```python
# compare the 4 models using a bar plot
```

```python
import matplotlib.pyplot as plt
models = ['knn','lg','dt','nb']
accuracy = [acc_knn,acc_lg,acc_dt,acc_nb]
plt.bar(models,accuracy,color=['green','blue','pink','brown'])
plt.xlabel("models")
plt.ylabel("Accuracy")
plt.show()
```

```python
import pandas as pd
import tkinter.messagebox as m
df=pd.read_csv("D:\\My Real Hero\\Haidar03.csv",names=['sl','sw','pl','pw','class'])
#  convert the last column to numeric
df['class']=df['class'].map({'Iris-setosa':0,"Iris-versicolor":1,"Iris-virginica":2})
# separate the dataframe in to input and output
X=df.drop(columns=['class'])  # input
Y=df['class']  #  output
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)

def knn():
    global acc_knn
    # create a model KNM   (k nearest neighbors model)
    from sklearn.neighbors import KNeighborsClassifier
    K=KNeighborsClassifier(n_neighbors=5)
    # When the model by using training data (80%)
    K.fit(x_train,y_train)
    # test the model
    y_pred_knn=K.predict(x_test)
    # find accuracy of model
    from sklearn.metrics import accuracy_score
    acc_knn=accuracy_score(y_test,y_pred_knn)
    acc_knn=round(acc_knn*100,2)
    m.showinfo(title="KNN",message="accuracy in knn is "+str(acc_knn)+"%")

def lg():
    global acc_lg
    global L
    # imlements logistic regression
    import warnings
    warnings.filterwarnings('ignore')
    from sklearn.linear_model import LogisticRegression
    L=LogisticRegression()
    L.fit(x_train,y_train)
    y_pred_lg= L.predict(x_test)
    from sklearn.metrics import accuracy_score
    acc_lg = accuracy_score(y_test,y_pred_lg)
    acc_lg=round(acc_lg*100,2)
    m.showinfo(title="LG",message="accuracy in LG is "+str(acc_lg)+"%")

def dt():
    global acc_dt
    # implement decision tree model
    from sklearn.tree import DecisionTreeClassifier
    D= DecisionTreeClassifier()
    D.fit(x_train,y_train)
    y_pred_dt=D.predict(x_test)
    from sklearn.metrics import accuracy_score
    acc_dt=accuracy_score(y_test,y_pred_dt)
    acc_dt=round(acc_dt*100,2)
    m.showinfo(title="DT",message="accuracy in DT is "+str(acc_dt)+"%")

def nb():
    global acc_nb
    # implement Naive nayes model
    from sklearn.naive_bayes import GaussianNB
    N=GaussianNB()
    N.fit(x_train,y_train)
    y_pred_nb=N.predict(x_test)
    from sklearn.metrics import accuracy_score
    acc_nb=accuracy_score(y_test,y_pred_nb)
    acc_nb=round(acc_nb*100,2)
    m.showinfo(title="NB",message="accuracy in NB is "+str(acc_nb)+"%")

def compare():
    import matplotlib.pyplot as plt
    models = ['knn','lg','dt','nb']
    accuracy = [acc_knn,acc_lg,acc_dt,acc_nb]
    plt.bar(models,accuracy,color=['green','blue','pink','brown'])
    plt.xlabel("models")
    plt.ylabel("Accuracy")
    plt.show()
    
def Submit():
    sl = float(Esl.get())
    sw = float(Esw.get())
    pl = float(Epl.get())
    pw = float(Epw.get())
    result = L.predict([[sl,sw,pl,pw]])
    if result[0]==0:
        flower = "SETOSA"
    elif result[0]==1:
        flower = "VERSICOLOR"
    else: 
        flower = "VIRGINICA"
    m.showinfo(title="IRIS",message=flower)
    
def Reset():
    Esl.delete(0, END)
    Esw.delete(0, END)
    Epl.delete(0, END)
    Epw.delete(0, END)

#####################################################################################
# implement the project using GUI...

from tkinter import *
import tkinter.messagebox as m
w = Tk()

# 1st row
L1=Label(w,text="IRIS FLOWER PROJECT",font=('arial',20,'bold'),bg='orange')
L1.grid(row=1,column=1,columnspan=4)

# 2nd row
Bknn=Button(w,text="KNN",font=('arial',20,'bold'),command=knn)
Blg=Button(w,text="LG",font=('arial',20,'bold'),command=lg)
Bdt=Button(w,text="DT",font=('arial',20,'bold'),command=dt)
Bnb=Button(w,text="NB",font=('arial',20,'bold'),command=nb)
Bknn.grid(row=2,column=1)
Blg.grid(row=2,column=2)
Bdt.grid(row=2,column=3)
Bnb.grid(row=2,column=4)

# 3rd row
Bcmp=Button(w,text="COMPARE",font=('arial',20,'bold'),command=compare)
Bcmp.grid(row=3, column=2,columnspan=2)

# 4th row
L2=Label(w,text="PREDICT FOR NEW FLOWER",font=('arial',17,'bold'),bg='orange')
L2.grid(row=4, column=1,columnspan=4)

# 5th row 
Lsl=Label(w,text="SL",font=('arial',20,'bold'))
Esl=Entry(w,font=('arial',20,'bold'),width=5)
Lsw=Label(w,text="SW",font=('arial',20,'bold'))
Esw=Entry(w,font=('arial',20,'bold'),width=5)
Lsl.grid(row=5,column=1)
Esl.grid(row=5,column=2)
Lsw.grid(row=5,column=3)
Esw.grid(row=5,column=4)

# 6th row
Lpl=Label(w,text="PL",font=('arial',20,'bold'))
Epl=Entry(w,font=('arial',20,'bold'),width=5)
Lpw=Label(w,text="PW",font=('arial',20,'bold'))
Epw=Entry(w,font=('arial',20,'bold'),width=5)
Lpl.grid(row=6,column=1)
Epl.grid(row=6,column=2)
Lpw.grid(row=6,column=3)
Epw.grid(row=6,column=4)

# 7th row
Bsub=Button(w,text="SUBMIT",font=('arial',15,'bold'),bg='green',command = Submit)
Bres=Button(w,text="RESET",font=('arial',15,'bold'),bg='red', command = Reset)
Bsub.grid(row=7,column=1,columnspan=2)
Bres.grid(row=7,column=3,columnspan=2)

w.mainloop()
```
